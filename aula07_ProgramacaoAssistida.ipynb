```python
# Initialize a list named 'numero' with integer values.
numero = [1,2,3]

# Calculate the sum of all elements in the 'numero' list using the built-in sum() function.
# The result is stored in the variable 'soma'.
soma = sum(numero)

# Print the calculated sum to the console using an f-string for formatted output.
print(f"A soma dos números é: {soma}")
```

```python
# Initialize a list named 'numeros' with integer values.
numeros = [1,3]

# Initialize a variable 'soma' to store the sum, starting at 0.
soma = 0

# Iterate through each number in the 'numeros' list.
for num in numeros:
  # Add the current number to the 'soma' variable.
  soma += num

# Print the final calculated sum to the console.
print(soma)
```
